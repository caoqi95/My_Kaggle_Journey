### Data Preprocessing

```python
import pandas as pd

## check the info of train data
train_data = pd.read_csv("./data/train.csv")
train_data.info()
train_data.head(10)
```

```python
## check the info of test data
test_data = pd.read_csv("./data/test.csv")
test_data.head(6)
```

```python
data_X = train_data.iloc[:, 1:]
data_y = train_data.iloc[:, 0]
X_test = test_data
```

```python
data_X.shape
```

```python
X_test.shape
```

```python
data_y.shape
```

```python
data_y
```

### Build a Model - Using KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(data_X, data_y)
```

### Test Model

```python
import numpy as np

pred = neigh.predict(X_test)
print(pred)

df = pd.DataFrame(pred)
df.index.name='ImageId'
df.index+=1
df.columns=['Label']
df.to_csv("submission_KNN.csv",header=True)
```
